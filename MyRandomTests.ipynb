```python
# Even and Odd number
print("")
listSize = 100

addOdd = 1
addEven = 0

if listSize % 2 != 0:
    addOdd = 0
    addEven = 1


print(f"List size {listSize}")
#evenList = list(range(-listSize + addEven, listSize, 2))
#oddList = list(range(-listSize + addOdd, listSize, 2))

evenList = list(range(0, listSize, 2))
oddList = list(range(1, listSize, 2))

print("")
print(f"evenList {len(evenList)}")
print(f"oddList {len(oddList)}")

print("")
print("")
print("Even List")
print("")
for number in evenList:
    print(f"Number {number}")


print("")
print("")
print("Odd List")
print("")
for number in oddList:
    print(f"Number {number}")


print("")
print("")
print("reverse Odd list minus Even list")
print("")
oddList.reverse()
count = 0
sumReverseOddList = 0
for number in oddList:
    print(f"Number {number-evenList[count]}")
    sumReverseOddList += number-evenList[count]
    count += 1
    
print("")
print(f"reverse list sum {sumReverseOddList}")

print("")
print("")
print("Primes to 100")
print("")
for number in oddList:
    if number % 3 == 0:
        continue
    elif number % 5 == 0:
        continue
    elif number % 7 == 0:
        continue
    print(f"Number {number}")
```
